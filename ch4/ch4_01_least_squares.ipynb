```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('headbrain.csv')
data
```

```python
plt.scatter(data.iloc[:,2],data.iloc[:,3])
plt.xlabel(data.columns[2])
plt.ylabel(data.columns[3])
plt.xlim([2500,5000])
plt.ylim([800,1800])
plt.grid(True)
plt.show()
```

```python
A = np.ones([len(data),2])
print(A.shape)
b = np.ones([len(data),1])
print(b.shape)
```

```python
A[:,1] = data.iloc[:,2]
b[:,0] = data.iloc[:,3]
```

```python
x =np.linalg.lstsq(A,b)
print('C=',x[0][0])
print('D=',x[0][1])
```

```python
head = np.arange(2500,5000,1)
print(head.size)
brain = x[0][0] + x[0][1]*head
print(brain.size)
```

```python
plt.scatter(data.iloc[:,2],data.iloc[:,3])
plt.plot(head,brain,'r')
plt.xlabel(data.columns[2])
plt.ylabel(data.columns[3])
plt.xlim([2500,5000])
plt.ylim([800,1800])
plt.grid(True)
plt.show()
```
